```python
import requests

def get_verification_code_from_api():
    url = "https://tools.dongvanfb.net/api/get_messages_oauth2"
    
    # Nhập chuỗi với định dạng email|pass|oauth|client_id
    nhap = input("Nhập thông tin theo định dạng 'email|pass|oauth|client_id': ")
    
    try:
        # Tách chuỗi dựa trên ký tự '|' và kiểm tra độ dài
        parts = nhap.split('|')
        if len(parts) < 4:
            print("Định dạng không đúng! Hãy nhập đủ 4 thành phần ngăn cách bằng '|'.")
            return None
        
        # Lấy email, oauth, và client_id
        email = parts[0]
        oauth = parts[2]
        client_id = parts[3]
        
        # Dữ liệu gửi lên API
        data = {
            "email": email,
            "refresh_token": oauth,
            "client_id": client_id
        }
        
        # Gửi yêu cầu POST
        response = requests.post(url, json=data)
        response.encoding = 'utf-8'  # Đảm bảo xử lý phản hồi đúng
        if response.status_code == 200:
            response_data = response.json()
            if "messages" in response_data:
                messages = response_data["messages"]
                # Nếu messages là list, nối thành chuỗi
                if isinstance(messages, list):
                    messages = " ".join(str(msg) for msg in messages)
                # Trích xuất mã xác minh
                if "Mã xác minh: " in messages and "bạn đang thực hiện đăng ký" in messages:
                    code = messages.split("Mã xác minh: ")[1].split(", bạn đang thực hiện đăng ký")[0]
                    return code
                else:
                    print("Không tìm thấy mã xác minh trong email.")
            else:
                print("Không tìm thấy trường 'messages' trong phản hồi.")
        else:
            print(f"API trả về lỗi. Mã trạng thái: {response.status_code}")
    except requests.exceptions.RequestException as e:
        print(f"Lỗi khi gọi API: {e}")
    except Exception as e:
        print(f"Lỗi khác: {e}")
    return None

# Sử dụng hàm
verification_code = get_verification_code_from_api()
if verification_code:
    print(f"Mã xác minh: {verification_code}")
else:
    print("Không lấy được mã xác minh.")
```
